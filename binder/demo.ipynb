```julia
using Dates
init_time = Dates.now()

cur_x = 10 .^ (0:0.1:1)
cur_y = rand(0.25:0.05:0.75, length(cur_x))
cur_z = rand(10)

@demo for a in 0:3, xlog = [true,false], title="title " .* string.(1:6)  
  scatter(
    cur_z, label="Scatter", color=4,
    xlabel="x", markersize=6
  )

  ylabel!("y")

  plot!(cur_x, cur_y, label="Plot", color=3, title=title)

  hline!([1,3] ./ 4, color=2)
  vline!(5, linewidth=4, alpha=0.4, linestyle=:dash)

  plot!(xlim=(1,15))

  ylims!(0,a)
  xscale!(xlog == true ? :log10 : :linear)
end

date2 = Dates.now()
println(
  "   ", round((Dates.now() - init_time) / Millisecond(1) * (1 / 1000), digits=6), " seconds"
)
```
